```python
# Срезы строк
# Если количество символов в сообщении чётное, то сообщение надо разделить ровно посередине и первую половину поместить в конец, а вторую — в начало.
# Если количество символов в сообщении нечётное, то каждый второй символ надо переместить в новую строку. После чего новую строку с перемещёнными символами приклеить в конец к первой.
# Пример входных данных (чётное количество символов): привет
# Пример ответа: ветпри

# Пример входных данных (нечётное количество символов): телефон
# Пример ответа: тлфнеео
```

```python
text = "телефон"
```

```python
length = len(text)
    
if length % 2 == 0:
    result = text[length//2:200] + text[0:length//2]
else:
    length % 2 == 1
    result = text[::2] + text[1::2]
    print(result)
```
